```python
df.rename(columns={'TotalAmount':'Sales'},inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["Region"]=df["Region"].fillna(df["Region"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["Product"]=df["Product"].fillna(df["Product"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["Quantity"]=pd.to_numeric(df["Quantity"])
```

```python
df["Quantity"].unique()
```

```python
df["Quantity"]=df["Quantity"].replace("five",'5')
```

```python
df.dtypes
```

```python
df["UnitPrice"].unique()
```

```python
df["UnitPrice"]=pd.to_numeric(df["UnitPrice"])
```

```python
df["UnitPrice"]=df["UnitPrice"].replace("unknown",'0')
```

```python
df["UnitPrice"].unique()
```

```python
df.dtypes
```

```python
df["OrderDate"]=pd.to_datetime(df["OrderDate"])
```

```python
df["UnitPrice"]=df["UnitPrice"].replace(0,df["UnitPrice"].mean())
```

```python
df.isnull().sum()
```

```python
df["Quantity"]=df["Quantity"].fillna(df["Quantity"].mean())
```

```python
df["UnitPrice"]=df["UnitPrice"].fillna(df["UnitPrice"].mean())
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
d1=df.groupby("Region")["Sales"].sum().reset_index()
d1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x="Region",y="Sales",data=d1,hue="Region")
plt.show()
```

```python
d2=df.groupby("Product")["Sales"].sum().reset_index()
d2
```

```python
sns.barplot(x="Product",y="Sales",data=d2,hue="Product")
plt.show()
```

```python
d3=df.groupby("Product")["UnitPrice"].mean().reset_index()
```

```python
sns.barplot(x="Product",y="UnitPrice",data=d3,hue="Product")
plt.show()
```

```python
d3=df.groupby("SalesPerson")["Quantity"].sum().reset_index()
```

```python
sns.barplot(x="SalesPerson",y="Quantity",data=d3,hue="SalesPerson")
plt.show()
```

```python
df["Month"]=df["OrderDate"].dt.month
d4=df.groupby("Month")["Sales"].sum().reset_index()
```

```python
sns.barplot(x="Month",y="Sales",data=d4,hue="Month")
plt.show()
```
